```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

Data Preprocessing


```python
# load the dataset to the pandas DataFrame
raw_mail_data = pd.read_csv('spamham.csv')
# replace the null values with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)), ' ')
```

```python
mail_data.shape
```

```python
mail_data.head() # sample of data
```

```python
# label spam mail as 0, and ham mail as 1
mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
# separate the data as text and label. x--> text; y--> label
x = mail_data['Message']
y = mail_data['Category']
```

```python
print(x)
print("......... ")
print(y)
```

Train Test Split

```python
# split the data as Train data and Test data
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8, test_size=0.2, random_state=3)
```

Feature Extraction

```python
# transform text data to feature vector that can be used as input to the SVM model using Tfidfvectorizer
# convert the text to lower letter
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase='True')
x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)

# covert y_train data and y_test data values in integers
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

Training the model--> **Support Vector Machine**

```python
# training the support vector machine model
model = LinearSVC()
model.fit(x_train_features, y_train)
```

Evaluation of the model

```python
# prediction of traing data
prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data is ',accuracy_on_training_data)
```

```python
# prediction of test data
prediction_on_test_data = model.predict(x_test_features)
accuracy_on_test_data = accuracy_score(y_test, prediction_on_test_data)
```

```python
print('Accuracy on test data is ',accuracy_on_test_data)
```

```python
# prediction on New mail
input_mail = [ "I'm gonna be home soon and i don't want to talk about this stuff anymore tonight, k? I've cried enough today."]
# convert text to feature vector
input_mail_features = feature_extraction.transform(input_mail)

# prediction
prediction = model.predict(input_mail_features)
print(prediction)

if (prediction [0]==1):
  print("Ham mail")
else:
  print("Spam mail")  
```
